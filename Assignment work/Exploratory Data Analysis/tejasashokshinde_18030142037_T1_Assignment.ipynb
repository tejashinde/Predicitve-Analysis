## Now that we have dropped the non relevant columns/features, we need to handle the other NaN's. Understood by the sum of null's in the columns (Age , Experience , Height , Pos , Team , BMI)

Handling Mechanism : 

 <table style="width:80%">
  <tr>
    <th>Column</th>
    <th>Strategy</th>
    <th>Description</th>  
  </tr>
  <tr>
    <td>Age</td>
    <td>Mean</td>
    <td>Taking a mean(Average) of all the ages </td>

  <tr>
    <td>Experience</td>
    <td>Mean</td>
    <td>Inspite of this column having categorical data, the elements are first converted to zeros , for instance 'R' means Rookie here and we can replace it with zero years of experience just for making it as a float, further calculating mean for the same. 
 
 As the dtype for the same is 'o'(Object) we need to convert it to float beafore taking the mean.</td>
  </tr>
  <tr>
    <td>Height</td>
    <td>Mean</td>
    <td>Taking a mode here will be slightly unfiar, as it is not necessary that if the frequence of a certain heigh is more, the NaN value will be it. So to be 'close to' accurate, mean is taken</td>
  </tr>
    <tr>
    <td>Weight</td>
    <td>Mean</td>
    <td>Taking a mode here will be slightly unfiar, as it is not necessary that if the frequence of a certain heigh is more, the NaN value will be it. So to be 'close to' accurate, mean is taken</td>
  </tr>
  <tr>
    <td>Position</td>
    <td>Categorical Codes</td>
    <td>Position being a categorical value, filling it's Null with a new defined category called "Not Assigned" and then converting these categories into the cat codes just so that we get the values in numbers.</td>
  </tr>
  <tr>
    <td>Team</td>
    <td>Converting NaN's to No Team Record</td>
    <td>"Not assigned" is a better value for the NaN's as while Plotting, the names would be explainatory rather than floats like numpy nan's</td>
  </tr>
  <tr>
    <td>BMI</td>
    <td>Calculation</td>
    <td>A Mean would be better than mode in  my opinion as Mode will give you the frequency making it 'not' necessary for the other players to have the same BMI count like others in majority. So Mean would not give an accurate value, but also it wouldn't converge from an accurate one</td>
  </tr>
</table> 

```python
# Age
dataframe['Age'].fillna(int(dataframe['Age'].mean()) , inplace = True)

# Height
dataframe['Height'].fillna(int(dataframe['Height'].mean()) , inplace = True)

# Height
dataframe['Weight'].fillna(int(dataframe['Weight'].mean()) , inplace = True)

# Experience
dataframe['Experience'].replace({'R' : 0.0} , inplace = True)
dataframe['Experience'].fillna(int(dataframe['Experience'].astype(float).mean()) , inplace = True)

# Position (Categorical) -> Converting the positions to category codes
dataframe.Pos.fillna("Not Assigned" , inplace = True)
# dataframe.Pos = pandas.Categorical(dataframe['Pos'] , dataframe.Pos.unique().tolist() , ordered = True)

# Team (Descriptive)
dataframe.Team.fillna("No Team Record" , inplace = True)

# BMI 
dataframe.BMI.fillna(dataframe.Weight / (((dataframe.Height)/100)**2) , inplace = True)
```

## Plotting the heat map of the correlation of a column to each other column

```python
corr = dataframe.corr()
seaborn.heatmap(corr , xticklabels = corr.columns.values,yticklabels = corr.columns.values , center = 1 , linewidths = 2 , annot = True)

print("Loading the graph.")
for iterator in tqdm(range(10)):
    time.sleep(0.2)
    
```

## Getting column pairs with high co-relation (Above 9.5)

```python
highCorrPairs = numpy.where(corr > 0.95)
highCorrPairs = [(corr.index[x], corr.columns[y]) for x, y in zip(*highCorrPairs) if x != y and x < y]
lowCorrPairs = numpy.where(corr < 0.2)
lowCorrPairs = [(corr.index[x], corr.columns[y]) for x, y in zip(*lowCorrPairs) if x != y and x < y]
```

```python
print("Column Pairs with high co-relation.")
print(pandas.DataFrame(highCorrPairs))

print("\n\n\n")

print("Column Pairs with low co-relation.")
print(pandas.DataFrame(lowCorrPairs))
```

```python
fig , axes = pyplot.subplots(nrows = int(len(highCorrPairs) / 2) , ncols = int(len(highCorrPairs) / 2))
dataframe[['PTS','FGM']].plot(ax = axes[0,0] , kind = 'box' , title = "Points and Field Goals Made")
dataframe[['PTS','FGA']].plot(ax = axes[0,1] , kind = 'box' , title = "Points and Field Foals Attempted")
dataframe[['FGM','FGA']].plot(ax = axes[0,2] , kind = 'box' , title = "Field Goals Made and Field Goals Attempted")
dataframe[['FGM','EFF']].plot(ax = axes[1,0] , kind = 'box' , title = "Field Goals Made and Efficiency")
dataframe[['3PM','3PM']].plot(ax = axes[1,1] , kind = 'box' , title = "3 pointers made and 3 pointers attempted")
dataframe[['FTM','FTA']].plot(ax = axes[1,2] , kind = 'box' , title = "Free Throws Made and Free Throws Attempted")
dataframe[['DREB','REB']].plot(ax = axes[2,0] , kind = 'box' , title = "Rebound and Defensive Rebound")
```

#### Replacing the Position characters to numeric values for further analysis

```python
defenseAnalysis.Pos.unique()
defenseAnalysis['Numeric Positions'] = defenseAnalysis['Pos'].replace({'PG' : 1 , 'PF' : 2 , 'SG' : 3 , 'SF' : 4 , 'C' : 5 , 'Not Assigned' : 0})
defenseAnalysis.set_index('Numeric Positions')
```
